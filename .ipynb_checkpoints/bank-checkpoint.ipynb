### 2.4 Thu giảm dữ liệu

Xem xét dữ liệu, ta thấy các cột đều có thể ảnh hưởng đến kết quả nên ta sẽ không xoá bất kỳ cột nào

### 2.5 Rời rạc hoá dữ liệu
Ta xem xét các giá trị của age

```python
df['age'].describe()
```

Ta thấy độ tuổi nhỏ nhất là 17 và lớn nhất là 98, cột age có độ lệch chuẩn cao nên ta cần phải thu gọn lại thành các đối tượng đặc trưng

```python
for dataset in [df]:
    dataset.loc[(dataset['age'] >= 17) & (dataset['age'] <= 30), 'age'] = 0
    dataset.loc[(dataset['age'] > 30) & (dataset['age'] <= 40), 'age'] = 1
    dataset.loc[(dataset['age'] > 40) & (dataset['age'] <= 65), 'age'] = 2
    dataset.loc[dataset['age'] > 65, 'age'] = 3
```
